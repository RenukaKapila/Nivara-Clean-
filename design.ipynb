```python
from pathlib import Path

ROOT = Path(".")
SRC = ROOT / "src"

files = {
    "src/lib/scam-lab.ts": r'''
export type PracticeScenarioFormat =
  | "sms"
  | "email"
  | "social"
  | "invoice"
  | "notice"
  | "image"
  | "pdf"

export type PracticeRiskLevel = "low" | "suspicious" | "high"

export type PracticeDifficulty = "Beginner" | "Intermediate" | "Advanced"

export interface PracticeWarningOption {
  id: string
  label: string
  why: string
}

export interface PracticeScenario {
  id: string
  format: PracticeScenarioFormat
  title: string
  hiddenCategory: string
  difficulty: PracticeDifficulty
  riskLevel: PracticeRiskLevel
  content: {
    sender?: string
    subject?: string
    body: string
    footer?: string
    link?: string
    amount?: string
    attachmentName?: string
  }
  warningOptions: PracticeWarningOption[]
  correctWarningIds: string[]
  explanation: string
  safeAction: string
  relatedLessonId: string
  relatedLessonLabel: string
}

const warningLibrary: Record<string, PracticeWarningOption> = {
  suspicious_link: {
    id: "suspicious_link",
    label: "Suspicious or unofficial link",
    why: "The link uses familiar words but does not clearly match the official website.",
  },
  urgency: {
    id: "urgency",
    label: "Urgent deadline or pressure",
    why: "Scammers create urgency so people act before verifying.",
  },
  threat: {
    id: "threat",
    label: "Threat or scary consequence",
    why: "Threats like suspension, arrest, or account closure are used to trigger panic.",
  },
  personal_info: {
    id: "personal_info",
    label: "Asks for personal information",
    why: "Unexpected requests for private details can lead to identity theft or account takeover.",
  },
  payment_request: {
    id: "payment_request",
    label: "Unexpected payment request",
    why: "Scammers often ask for money through fees, fines, deposits, or fake balances.",
  },
  gift_cards: {
    id: "gift_cards",
    label: "Gift card request",
    why: "Gift cards are a major warning sign because they are hard to recover once shared.",
  },
  secrecy: {
    id: "secrecy",
    label: "Secrecy request",
    why: "Scammers isolate people by telling them not to tell family, coworkers, banks, or support.",
  },
  avoid_official: {
    id: "avoid_official",
    label: "Avoids official channels",
    why: "A message that says not to call support, your bank, or the real office is suspicious.",
  },
  verification_code: {
    id: "verification_code",
    label: "Verification code request",
    why: "Verification codes are for you only. Sharing them can give someone access to your account.",
  },
  too_good: {
    id: "too_good",
    label: "Too-good-to-be-true offer",
    why: "Huge rewards, easy money, or guaranteed returns are common scam hooks.",
  },
  unusual_payment: {
    id: "unusual_payment",
    label: "Unusual payment method",
    why: "Zelle, Cash App, crypto, wires, and gift cards are often hard to reverse.",
  },
  safe_official: {
    id: "safe_official",
    label: "Uses official verification steps",
    why: "A safer message tells you to use official apps, websites, or known phone numbers.",
  },
  normal_context: {
    id: "normal_context",
    label: "Normal expected context",
    why: "Not every reminder is suspicious. Expected messages with official verification language can be low risk.",
  },
}

interface ScenarioTemplate {
  category: string
  format: PracticeScenarioFormat
  difficulty: PracticeDifficulty
  riskLevel: PracticeRiskLevel
  relatedLessonId: string
  relatedLessonLabel: string
  subjects: string[]
  senders: string[]
  bodies: string[]
  links?: string[]
  amounts?: string[]
  correctWarningIds: string[]
  explanation: string
  safeAction: string
}

const templates: ScenarioTemplate[] = [
  {
    category: "Bank Impersonation",
    format: "sms",
    difficulty: "Beginner",
    riskLevel: "high",
    relatedLessonId: "phishing-fake-websites",
    relatedLessonLabel: "Phishing & Fake Websites",
    subjects: ["Bank Security Alert"],
    senders: ["Chase Security", "Bank Fraud Center", "Card Services"],
    bodies: [
      "A new device was added to your online banking profile. Verify now to prevent temporary suspension.",
      "Unusual login attempt detected. Confirm your identity within 2 hours to avoid account closure.",
      "Your debit card has been restricted. Use the secure review link to restore access.",
    ],
    links: [
      "https://chase-secure-login.accountreview-center.com",
      "https://bank-verify-session.secure-reviewhub.com",
      "https://cardservices-alert.identity-confirm.net",
    ],
    correctWarningIds: ["suspicious_link", "urgency", "threat", "personal_info"],
    explanation:
      "This scenario uses a fake bank-style alert, urgency, and a suspicious link to push the user into a fake login or identity verification page.",
    safeAction:
      "Do not click the link. Open the official bank app or call the number on the back of the card.",
  },
  {
    category: "Package Delivery Scam",
    format: "sms",
    difficulty: "Beginner",
    riskLevel: "high",
    relatedLessonId: "phishing-fake-websites",
    relatedLessonLabel: "Phishing & Fake Websites",
    subjects: ["Delivery Notice"],
    senders: ["USPS Notice", "Delivery Support", "Parcel Update"],
    bodies: [
      "Your package cannot be delivered because the address is incomplete. A small redelivery fee is required today.",
      "Final delivery attempt failed. Confirm your address and payment details to reschedule delivery.",
      "Your parcel is on hold. Update delivery information now to avoid return to sender.",
    ],
    links: [
      "https://usps-redelivery-fee.verify-address-support.com",
      "https://parcel-reschedule.delivery-helpcenter.net",
      "https://tracking-update.addressconfirm-service.com",
    ],
    amounts: ["$0.30", "$1.99", "$2.99"],
    correctWarningIds: ["suspicious_link", "payment_request", "urgency", "personal_info"],
    explanation:
      "This delivery message uses a tiny fee, a suspicious link, and address/payment pressure to collect personal or card information.",
    safeAction:
      "Track packages only through the official carrier website or the store where you ordered.",
  },
  {
    category: "Fake Job Offer",
    format: "email",
    difficulty: "Intermediate",
    riskLevel: "high",
    relatedLessonId: "payment-money-scams",
    relatedLessonLabel: "Payment & Money Scams",
    subjects: ["Remote Job Offer", "Work From Home Position", "Immediate Hiring Notice"],
    senders: ["Recruiting Team", "HR Department", "Talent Coordinator"],
    bodies: [
      "We reviewed your profile and selected you for a remote data entry role paying $38/hour. No interview is required. To receive your laptop, submit the equipment processing fee.",
      "Congratulations, you are approved for a remote assistant role. Training starts today after your refundable onboarding deposit is paid.",
      "Your resume was shortlisted. Please pay the background verification fee through Zelle or Cash App to continue.",
    ],
    amounts: ["$65", "$85", "$120"],
    correctWarningIds: ["too_good", "payment_request", "unusual_payment", "personal_info"],
    explanation:
      "This job offer uses high pay, fast hiring, and an upfront fee. Real employers usually do not require applicants to pay to receive equipment or start work.",
    safeAction:
      "Do not pay. Verify the company through its official careers page and trusted contact information.",
  },
  {
    category: "Family Emergency Scam",
    format: "sms",
    difficulty: "Intermediate",
    riskLevel: "high",
    relatedLessonId: "social-engineering",
    relatedLessonLabel: "Social Engineering",
    subjects: ["Emergency Message"],
    senders: ["Unknown Number", "New Phone", "Private Number"],
    bodies: [
      "Mom, don’t call me right now. I’m in trouble and using someone else’s phone. Please send money for the lawyer and don’t tell anyone.",
      "I had an accident and need help before they let me leave. Please buy Apple gift cards and text me the codes.",
      "Please don’t call. I’m scared. I need money urgently and I can’t explain right now.",
    ],
    amounts: ["$400", "$700", "$900"],
    correctWarningIds: ["urgency", "secrecy", "gift_cards", "avoid_official"],
    explanation:
      "This scenario uses fear, secrecy, and a family emergency story to make the user act before verifying.",
    safeAction:
      "Call the person using their known number or contact another trusted family member before sending anything.",
  },
  {
    category: "Government Impersonation",
    format: "notice",
    difficulty: "Intermediate",
    riskLevel: "high",
    relatedLessonId: "social-engineering",
    relatedLessonLabel: "Social Engineering",
    subjects: ["Final Government Notice", "Case Review Required", "Legal Warning"],
    senders: ["Federal Review Office", "Social Security Department", "Tax Compliance Unit"],
    bodies: [
      "Your Social Security number has been linked to suspicious activity. Immediate verification is required to avoid suspension.",
      "A legal case has been opened under your name. Failure to respond today may result in enforcement action.",
      "Your tax file requires immediate payment to avoid penalties and legal escalation.",
    ],
    links: [
      "https://ssa-case-review.identity-verificationcenter.com",
      "https://federal-legalnotice.secure-response.net",
      "https://tax-compliance-pay.reviewdesk.org",
    ],
    correctWarningIds: ["threat", "urgency", "suspicious_link", "payment_request"],
    explanation:
      "This message impersonates authority and uses threats to pressure the user. Real government agencies usually do not demand immediate action through random links.",
    safeAction:
      "Do not use the link. Visit the official agency website directly or call an official published number.",
  },
  {
    category: "Subscription Refund Scam",
    format: "email",
    difficulty: "Intermediate",
    riskLevel: "high",
    relatedLessonId: "payment-money-scams",
    relatedLessonLabel: "Payment & Money Scams",
    subjects: ["Renewal Invoice", "Subscription Charge Notice", "Refund Case Opened"],
    senders: ["Billing Support", "Security Renewal", "Subscription Desk"],
    bodies: [
      "Your account has been charged for annual device protection. If this was not authorized, call immediately for cancellation and refund.",
      "Your subscription renewal is complete. To cancel the charge, contact our billing department today.",
      "A refund case has been opened. Do not contact your bank while the cancellation is being processed.",
    ],
    amounts: ["$299.99", "$399.99", "$499.99"],
    correctWarningIds: ["urgency", "avoid_official", "payment_request", "threat"],
    explanation:
      "Refund scams often use fake invoices and urgent cancellation language to move victims onto a phone call controlled by the scammer.",
    safeAction:
      "Do not call the number in the message. Check your real account or contact the company through its official website.",
  },
  {
    category: "Wrong-Number Investment Scam",
    format: "social",
    difficulty: "Advanced",
    riskLevel: "high",
    relatedLessonId: "social-engineering",
    relatedLessonLabel: "Social Engineering",
    subjects: ["Wrong Number Chat"],
    senders: ["Unknown Contact", "New Friend", "Investment Mentor"],
    bodies: [
      "Sorry, is this Priya? My assistant gave me the wrong number. You seem kind though. I’m investing in short-term crypto contracts and my mentor helped me earn a lot this week.",
      "I thought this was my friend. Anyway, you seem nice. I can show you an invite-only trading platform if you want.",
      "Wrong number, sorry. By the way, I work with a private investment group. The entry window closes tonight.",
    ],
    amounts: ["$4,800", "$7,200", "$12,500"],
    correctWarningIds: ["too_good", "urgency", "unusual_payment", "personal_info"],
    explanation:
      "Wrong-number scams often start casually, build trust, and later introduce crypto or investment opportunities.",
    safeAction:
      "Do not invest through a stranger, wrong-number message, or invite-only platform.",
  },
  {
    category: "MFA Code Theft",
    format: "sms",
    difficulty: "Intermediate",
    riskLevel: "high",
    relatedLessonId: "passwords-mfa-account-safety",
    relatedLessonLabel: "Passwords, MFA & Account Safety",
    subjects: ["Security Code"],
    senders: ["Support Agent", "Account Help", "Verification Desk"],
    bodies: [
      "We are helping secure your account. Please read the 6-digit verification code you just received so we can confirm it is you.",
      "Your account recovery is almost complete. Reply with the code sent to your phone.",
      "To cancel the suspicious login, send us the verification code immediately.",
    ],
    correctWarningIds: ["verification_code", "urgency", "personal_info", "avoid_official"],
    explanation:
      "A verification code request is a major warning sign. Codes are used to access accounts and should never be shared.",
    safeAction:
      "Do not share the code. Go directly to the official account security page and review recent activity.",
  },
  {
    category: "Device Remote Access Scam",
    format: "notice",
    difficulty: "Intermediate",
    riskLevel: "high",
    relatedLessonId: "device-app-safety",
    relatedLessonLabel: "Device & App Safety",
    subjects: ["Device Warning", "Virus Alert", "Support Required"],
    senders: ["System Security", "Tech Support", "Device Protection Center"],
    bodies: [
      "Your computer is infected. Call support and install the remote support tool so a technician can remove the threat.",
      "Your device has been compromised. Download the security repair app now.",
      "We detected unauthorized activity. A technician must connect to your screen immediately.",
    ],
    links: [
      "https://secure-device-repair.supportdownload.net",
      "https://system-cleanup.remotehelpcenter.com",
      "https://quickfix-securitytool.download",
    ],
    correctWarningIds: ["suspicious_link", "urgency", "threat", "avoid_official"],
    explanation:
      "Remote access scams try to convince users to install software that gives scammers control of the device.",
    safeAction:
      "Do not install tools from popups or unknown callers. Use official device support or trusted security software.",
  },
  {
    category: "Identity Information Harvesting",
    format: "pdf",
    difficulty: "Advanced",
    riskLevel: "high",
    relatedLessonId: "identity-protection",
    relatedLessonLabel: "Identity Protection",
    subjects: ["Identity Verification Form", "Employment Verification PDF", "Benefit Review Document"],
    senders: ["Document Review", "Benefits Office", "Hiring Verification"],
    bodies: [
      "Complete the attached form with your SSN, date of birth, address, bank name, and photo ID to continue processing.",
      "Your application is paused until identity documents are uploaded through the attached form.",
      "To avoid delays, return this verification document with full personal details today.",
    ],
    correctWarningIds: ["personal_info", "urgency", "suspicious_link", "avoid_official"],
    explanation:
      "This scenario asks for highly sensitive identity information through an unexpected document or form.",
    safeAction:
      "Verify the organization through an official channel before sending identity documents or SSN information.",
  },
  {
    category: "Safe Official Reminder",
    format: "email",
    difficulty: "Beginner",
    riskLevel: "low",
    relatedLessonId: "safe-verification-habits",
    relatedLessonLabel: "Safe Verification Habits",
    subjects: ["Appointment Reminder", "Tuition Reminder", "Account Notice"],
    senders: ["Official Office", "Student Accounts", "Clinic Reminder"],
    bodies: [
      "This is a reminder that your appointment is scheduled for Tuesday. If you need help, call the office using the number listed on our official website.",
      "Your tuition deadline is coming up. Please log in through your official student portal to view your balance.",
      "A document is ready in your official account portal. Please open the app or visit the official website directly.",
    ],
    correctWarningIds: ["safe_official", "normal_context"],
    explanation:
      "This message uses safer verification language by telling the user to use an official portal, app, website, or known contact method.",
    safeAction:
      "If expected, use the official portal or known phone number. Do not use random links from suspicious messages.",
  },
]

const decoyWarnings = Object.values(warningLibrary)

function getOptions(correctIds: string[]): PracticeWarningOption[] {
  const correct = correctIds.map((id) => warningLibrary[id]).filter(Boolean)
  const decoys = decoyWarnings
    .filter((item) => !correctIds.includes(item.id))
    .slice(0, 4)

  return [...correct, ...decoys].sort((a, b) => a.label.localeCompare(b.label))
}

function titleFromCategory(category: string, index: number): string {
  const labels = [
    "Practice Scenario",
    "Review This Situation",
    "What Would You Notice?",
    "Pause Before You Trust",
    "Spot the Warning Signs",
  ]
  return `${labels[index % labels.length]} ${index + 1}`
}

export const practiceScenarios: PracticeScenario[] = templates.flatMap((template, templateIndex) => {
  const results: PracticeScenario[] = []

  template.bodies.forEach((body, bodyIndex) => {
    const senders = template.senders.length ? template.senders : ["Unknown Sender"]
    const subjects = template.subjects.length ? template.subjects : ["Practice Scenario"]
    const links = template.links && template.links.length ? template.links : [undefined]
    const amounts = template.amounts && template.amounts.length ? template.amounts : [undefined]

    senders.forEach((sender, senderIndex) => {
      links.forEach((link, linkIndex) => {
        amounts.forEach((amount, amountIndex) => {
          const scenarioIndex =
            templateIndex * 1000 + bodyIndex * 100 + senderIndex * 10 + linkIndex + amountIndex

          const amountText = amount ? ` Amount mentioned: ${amount}.` : ""
          const linkText = link ? ` Link shown: ${link}` : ""

          results.push({
            id: `scenario-${templateIndex}-${bodyIndex}-${senderIndex}-${linkIndex}-${amountIndex}`,
            format: template.format,
            title: titleFromCategory(template.category, scenarioIndex),
            hiddenCategory: template.category,
            difficulty: template.difficulty,
            riskLevel: template.riskLevel,
            content: {
              sender,
              subject: subjects[bodyIndex % subjects.length],
              body: `${body}${amountText}${linkText}`,
              link,
              amount,
              attachmentName:
                template.format === "pdf"
                  ? "verification-document.pdf"
                  : template.format === "invoice"
                    ? "invoice-review.pdf"
                    : undefined,
            },
            warningOptions: getOptions(template.correctWarningIds),
            correctWarningIds: template.correctWarningIds,
            explanation: template.explanation,
            safeAction: template.safeAction,
            relatedLessonId: template.relatedLessonId,
            relatedLessonLabel: template.relatedLessonLabel,
          })
        })
      })
    })
  })

  return results
})

export function getPracticeScenarioCount() {
  return practiceScenarios.length
}
''',

    "src/lib/scam-lab-engine.ts": r'''
import { practiceScenarios, type PracticeScenario } from "./scam-lab"

const STORAGE_KEY = "nivara_practice_completed_scenarios_v1"

function canUseStorage() {
  return typeof window !== "undefined" && typeof window.localStorage !== "undefined"
}

export function getCompletedScenarioIds(): string[] {
  if (!canUseStorage()) return []

  try {
    const raw = window.localStorage.getItem(STORAGE_KEY)
    if (!raw) return []

    const parsed = JSON.parse(raw)
    return Array.isArray(parsed) ? parsed.filter((item) => typeof item === "string") : []
  } catch {
    return []
  }
}

export function saveCompletedScenarioId(id: string) {
  if (!canUseStorage()) return

  const completed = new Set(getCompletedScenarioIds())
  completed.add(id)

  window.localStorage.setItem(STORAGE_KEY, JSON.stringify([...completed]))
}

export function resetCompletedScenarios() {
  if (!canUseStorage()) return
  window.localStorage.removeItem(STORAGE_KEY)
}

export function getRemainingScenarios(): PracticeScenario[] {
  const completed = new Set(getCompletedScenarioIds())
  return practiceScenarios.filter((scenario) => !completed.has(scenario.id))
}

export function getRandomScenario(): PracticeScenario | null {
  const remaining = getRemainingScenarios()
  if (remaining.length === 0) return null

  const index = Math.floor(Math.random() * remaining.length)
  return remaining[index]
}

export function getPracticeProgress() {
  const completedCount = getCompletedScenarioIds().length
  const totalCount = practiceScenarios.length
  const remainingCount = Math.max(totalCount - completedCount, 0)
  const percent = totalCount > 0 ? Math.round((completedCount / totalCount) * 100) : 0

  return {
    completedCount,
    totalCount,
    remainingCount,
    percent,
  }
}
''',

    "src/pages/ScamLab.tsx": r'''
import { useEffect, useMemo, useState } from "react"
import { Link } from "react-router-dom"
import {
  ArrowRight,
  BookOpen,
  CheckCircle,
  FileText,
  FlaskConical,
  ImageIcon,
  Mail,
  MessageSquare,
  RefreshCcw,
  RotateCcw,
  Shield,
  Sparkles,
  XCircle,
} from "lucide-react"
import { Button } from "@/components/ui/button"
import { Badge } from "@/components/ui/badge"
import { Card, CardContent } from "@/components/ui/card"
import {
  getPracticeProgress,
  getRandomScenario,
  resetCompletedScenarios,
  saveCompletedScenarioId,
} from "@/lib/scam-lab-engine"
import type { PracticeRiskLevel, PracticeScenario, PracticeScenarioFormat } from "@/lib/scam-lab"

const riskOptions: Array<{ value: PracticeRiskLevel; label: string; helper: string }> = [
  { value: "low", label: "Low Risk", helper: "Looks mostly normal or expected" },
  { value: "suspicious", label: "Suspicious", helper: "Some warning signs are present" },
  { value: "high", label: "High Risk", helper: "Multiple strong warning signs" },
]

const formatIcons: Record<PracticeScenarioFormat, typeof MessageSquare> = {
  sms: MessageSquare,
  email: Mail,
  social: MessageSquare,
  invoice: FileText,
  notice: FileText,
  image: ImageIcon,
  pdf: FileText,
}

const formatLabels: Record<PracticeScenarioFormat, string> = {
  sms: "Text Message",
  email: "Email",
  social: "Social Message",
  invoice: "Invoice Preview",
  notice: "Notice",
  image: "Image-Style Example",
  pdf: "PDF Preview",
}

function ScenarioMockup({ scenario }: { scenario: PracticeScenario }) {
  const Icon = formatIcons[scenario.format] || MessageSquare

  return (
    <div className="rounded-2xl border border-border bg-white shadow-sm overflow-hidden">
      <div className="flex items-center justify-between border-b border-border bg-muted/30 px-4 py-3">
        <div className="flex items-center gap-2">
          <div className="flex h-8 w-8 items-center justify-center rounded-lg bg-brand-blue-light text-brand-blue">
            <Icon className="h-4 w-4" />
          </div>
          <div>
            <p className="text-xs font-bold text-foreground">{formatLabels[scenario.format]}</p>
            <p className="text-xs text-muted-foreground">Review the situation carefully</p>
          </div>
        </div>

        <Badge variant="outline" className="text-xs">
          {scenario.difficulty}
        </Badge>
      </div>

      <div className="p-4">
        {scenario.content.sender && (
          <p className="text-xs text-muted-foreground mb-1">
            From: <span className="font-semibold text-foreground">{scenario.content.sender}</span>
          </p>
        )}

        {scenario.content.subject && (
          <p className="text-sm font-bold text-foreground mb-3">{scenario.content.subject}</p>
        )}

        <div
          className={`rounded-2xl border p-4 ${
            scenario.format === "sms" || scenario.format === "social"
              ? "border-blue-200 bg-blue-50/60"
              : scenario.format === "pdf" || scenario.format === "invoice"
                ? "border-amber-200 bg-amber-50/60"
                : "border-border bg-muted/20"
          }`}
        >
          <p className="text-sm leading-relaxed text-foreground whitespace-pre-wrap">
            {scenario.content.body}
          </p>
        </div>

        {scenario.content.attachmentName && (
          <div className="mt-3 rounded-xl border border-border bg-muted/20 p-3 flex items-center gap-2">
            <FileText className="h-4 w-4 text-muted-foreground" />
            <span className="text-sm text-foreground">{scenario.content.attachmentName}</span>
          </div>
        )}

        {scenario.content.footer && (
          <p className="mt-3 text-xs text-muted-foreground">{scenario.content.footer}</p>
        )}
      </div>
    </div>
  )
}

function getScore({
  correctWarningIds,
  selectedWarningIds,
  selectedRisk,
  actualRisk,
}: {
  correctWarningIds: string[]
  selectedWarningIds: string[]
  selectedRisk: PracticeRiskLevel | null
  actualRisk: PracticeRiskLevel
}) {
  const correctSet = new Set(correctWarningIds)
  const selectedSet = new Set(selectedWarningIds)

  const correctSelected = [...selectedSet].filter((id) => correctSet.has(id)).length
  const incorrectSelected = [...selectedSet].filter((id) => !correctSet.has(id)).length
  const missed = correctWarningIds.length - correctSelected

  const warningScore =
    correctWarningIds.length > 0
      ? Math.max(0, Math.round(((correctSelected - incorrectSelected * 0.5) / correctWarningIds.length) * 70))
      : 0

  const riskScore = selectedRisk === actualRisk ? 30 : 0
  const total = Math.max(0, Math.min(100, warningScore + riskScore))

  return {
    total,
    correctSelected,
    incorrectSelected,
    missed,
    riskCorrect: selectedRisk === actualRisk,
  }
}

export function ScamLab() {
  const [scenario, setScenario] = useState<PracticeScenario | null>(null)
  const [selectedWarningIds, setSelectedWarningIds] = useState<string[]>([])
  const [selectedRisk, setSelectedRisk] = useState<PracticeRiskLevel | null>(null)
  const [submitted, setSubmitted] = useState(false)
  const [progress, setProgress] = useState(getPracticeProgress())

  function loadNewScenario() {
    const next = getRandomScenario()
    setScenario(next)
    setSelectedWarningIds([])
    setSelectedRisk(null)
    setSubmitted(false)
    setProgress(getPracticeProgress())
  }

  useEffect(() => {
    loadNewScenario()
  }, [])

  const score = useMemo(() => {
    if (!scenario) return null

    return getScore({
      correctWarningIds: scenario.correctWarningIds,
      selectedWarningIds,
      selectedRisk,
      actualRisk: scenario.riskLevel,
    })
  }, [scenario, selectedWarningIds, selectedRisk])

  function toggleWarning(id: string) {
    if (submitted) return

    setSelectedWarningIds((prev) =>
      prev.includes(id) ? prev.filter((item) => item !== id) : [...prev, id],
    )
  }

  function handleSubmit() {
    if (!scenario) return

    setSubmitted(true)
    saveCompletedScenarioId(scenario.id)
    setProgress(getPracticeProgress())
  }

  function handleResetAll() {
    resetCompletedScenarios()
    loadNewScenario()
  }

  if (!scenario) {
    return (
      <div className="min-h-screen bg-white">
        <div className="mx-auto max-w-2xl px-4 sm:px-6 lg:px-8 py-12">
          <div className="text-center">
            <div className="mx-auto mb-4 flex h-12 w-12 items-center justify-center rounded-xl bg-brand-blue text-white shadow-md">
              <FlaskConical className="h-6 w-6" />
            </div>

            <h1 className="text-3xl font-bold tracking-tight text-foreground">
              Practice Lab Complete
            </h1>

            <p className="mt-3 text-muted-foreground leading-relaxed">
              You have completed every available scenario in this practice set. Reset practice to start again.
            </p>

            <Button onClick={handleResetAll} className="mt-6 bg-brand-blue hover:bg-brand-blue/90 text-white gap-2">
              <RotateCcw className="h-4 w-4" />
              Reset Practice
            </Button>
          </div>
        </div>
      </div>
    )
  }

  const canSubmit = selectedWarningIds.length > 0 && selectedRisk !== null
  const correctSet = new Set(scenario.correctWarningIds)

  return (
    <div className="min-h-screen bg-white">
      <div className="mx-auto max-w-3xl px-4 sm:px-6 lg:px-8 py-10">
        <div className="mb-8 text-center">
          <div className="mx-auto mb-4 flex h-12 w-12 items-center justify-center rounded-xl bg-brand-blue text-white shadow-md">
            <FlaskConical className="h-6 w-6" />
          </div>

          <Badge className="mb-3 bg-brand-blue-light text-brand-blue border border-brand-blue/20">
            Random Practice
          </Badge>

          <h1 className="text-3xl font-bold tracking-tight text-foreground">Practice Lab</h1>

          <p className="mt-3 text-sm text-muted-foreground leading-relaxed max-w-xl mx-auto">
            Review realistic scenarios without seeing the category first. Spot the warning signs, choose the risk level, then learn the pattern.
          </p>
        </div>

        <Card className="mb-6 border-border shadow-sm">
          <CardContent className="p-5">
            <div className="flex items-center justify-between gap-4 mb-2">
              <div>
                <p className="text-xs font-semibold text-muted-foreground">Practice progress</p>
                <p className="text-sm font-bold text-foreground">
                  {progress.completedCount} completed · {progress.remainingCount} remaining
                </p>
              </div>

              <Button
                variant="outline"
                size="sm"
                onClick={handleResetAll}
                className="gap-2 border-border"
              >
                <RefreshCcw className="h-4 w-4" />
                Reset
              </Button>
            </div>

            <div className="h-2 rounded-full bg-muted overflow-hidden">
              <div
                className="h-full rounded-full bg-gradient-to-r from-brand-blue via-brand-teal to-emerald-400 transition-all duration-300"
                style={{ width: `${progress.percent}%` }}
              />
            </div>
          </CardContent>
        </Card>

        <div className="mb-6">
          <ScenarioMockup scenario={scenario} />
        </div>

        <Card className="mb-6 border-border shadow-sm">
          <CardContent className="p-5">
            <div className="mb-4">
              <p className="text-xs font-bold uppercase tracking-wide text-red-700">
                Step 1
              </p>
              <h2 className="text-lg font-bold text-foreground">
                What warning signs do you notice?
              </h2>
              <p className="text-sm text-muted-foreground">
                Select every sign that applies. Some options may be decoys.
              </p>
            </div>

            <div className="grid gap-2 sm:grid-cols-2">
              {scenario.warningOptions.map((option) => {
                const selected = selectedWarningIds.includes(option.id)

                return (
                  <button
                    key={option.id}
                    type="button"
                    onClick={() => toggleWarning(option.id)}
                    disabled={submitted}
                    className={`rounded-xl border p-3 text-left text-sm transition-all ${
                      selected
                        ? "border-brand-blue bg-brand-blue-light text-brand-blue"
                        : "border-border bg-white hover:bg-muted/30"
                    }`}
                  >
                    <div className="flex items-start gap-2">
                      {selected ? (
                        <CheckCircle className="h-4 w-4 mt-0.5 shrink-0" />
                      ) : (
                        <div className="h-4 w-4 rounded-full border-2 border-current mt-0.5 shrink-0" />
                      )}
                      <span className="font-medium">{option.label}</span>
                    </div>
                  </button>
                )
              })}
            </div>
          </CardContent>
        </Card>

        <Card className="mb-6 border-border shadow-sm">
          <CardContent className="p-5">
            <div className="mb-4">
              <p className="text-xs font-bold uppercase tracking-wide text-amber-700">
                Step 2
              </p>
              <h2 className="text-lg font-bold text-foreground">
                What risk level would you choose?
              </h2>
              <p className="text-sm text-muted-foreground">
                This helps you practice judgment, not just memorization.
              </p>
            </div>

            <div className="grid gap-2 sm:grid-cols-3">
              {riskOptions.map((option) => {
                const selected = selectedRisk === option.value

                return (
                  <button
                    key={option.value}
                    type="button"
                    onClick={() => !submitted && setSelectedRisk(option.value)}
                    disabled={submitted}
                    className={`rounded-xl border p-4 text-left transition-all ${
                      selected
                        ? "border-amber-300 bg-amber-50 text-amber-800"
                        : "border-border bg-white hover:bg-muted/30"
                    }`}
                  >
                    <p className="text-sm font-bold">{option.label}</p>
                    <p className="mt-1 text-xs text-muted-foreground">{option.helper}</p>
                  </button>
                )
              })}
            </div>
          </CardContent>
        </Card>

        {!submitted ? (
          <Button
            onClick={handleSubmit}
            disabled={!canSubmit}
            className="w-full bg-brand-blue hover:bg-brand-blue/90 text-white gap-2 mb-8"
          >
            Check My Answer
            <ArrowRight className="h-4 w-4" />
          </Button>
        ) : (
          <Card className="mb-8 border-brand-blue/20 bg-brand-blue-light/20 shadow-sm">
            <CardContent className="p-5">
              <div className="text-center mb-5">
                <Sparkles className="mx-auto mb-2 h-7 w-7 text-brand-blue" />
                <p className="text-3xl font-extrabold text-brand-blue">{score?.total}%</p>
                <p className="text-sm text-muted-foreground">
                  {score?.total === 100
                    ? "Excellent. You spotted the pattern clearly."
                    : score && score.total >= 70
                      ? "Good work. Review the missed details below."
                      : "Good practice. This one had some tricky signals."}
                </p>
              </div>

              <div className="grid gap-3 sm:grid-cols-2 mb-5">
                <div className="rounded-xl border border-border bg-white p-4">
                  <p className="text-xs font-bold text-muted-foreground uppercase tracking-wide">
                    Scenario Type
                  </p>
                  <p className="text-sm font-bold text-foreground mt-1">
                    {scenario.hiddenCategory}
                  </p>
                </div>

                <div className="rounded-xl border border-border bg-white p-4">
                  <p className="text-xs font-bold text-muted-foreground uppercase tracking-wide">
                    Risk Level
                  </p>
                  <p className="text-sm font-bold text-foreground mt-1">
                    {scenario.riskLevel.toUpperCase()}
                  </p>
                </div>
              </div>

              <div className="space-y-2 mb-5">
                {scenario.warningOptions.map((option) => {
                  const selected = selectedWarningIds.includes(option.id)
                  const correct = correctSet.has(option.id)

                  if (!selected && !correct) return null

                  return (
                    <div
                      key={option.id}
                      className={`rounded-xl border p-3 ${
                        correct
                          ? "border-emerald-200 bg-emerald-50"
                          : "border-red-200 bg-red-50"
                      }`}
                    >
                      <div className="flex items-start gap-2">
                        {correct ? (
                          <CheckCircle className="h-4 w-4 text-emerald-600 mt-0.5 shrink-0" />
                        ) : (
                          <XCircle className="h-4 w-4 text-red-600 mt-0.5 shrink-0" />
                        )}
                        <div>
                          <p className="text-sm font-bold text-foreground">
                            {option.label}
                          </p>
                          <p className="mt-1 text-xs text-muted-foreground leading-relaxed">
                            {option.why}
                          </p>
                        </div>
                      </div>
                    </div>
                  )
                })}
              </div>

              <div className="rounded-xl border border-border bg-white p-4 mb-4">
                <p className="text-xs font-bold text-muted-foreground uppercase tracking-wide">
                  Explanation
                </p>
                <p className="mt-2 text-sm text-foreground leading-relaxed">
                  {scenario.explanation}
                </p>
              </div>

              <div className="rounded-xl border border-emerald-200 bg-emerald-50 p-4 mb-5">
                <div className="flex items-start gap-2">
                  <Shield className="h-4 w-4 text-emerald-600 mt-0.5 shrink-0" />
                  <div>
                    <p className="text-sm font-bold text-emerald-700">Safer move</p>
                    <p className="mt-1 text-sm text-foreground leading-relaxed">
                      {scenario.safeAction}
                    </p>
                  </div>
                </div>
              </div>

              <div className="flex flex-col sm:flex-row gap-2">
                <Link to={`/learning/${scenario.relatedLessonId}`} className="flex-1">
                  <Button variant="outline" className="w-full gap-2 border-brand-blue/30 text-brand-blue hover:bg-brand-blue-light">
                    <BookOpen className="h-4 w-4" />
                    Learn This Pattern
                  </Button>
                </Link>

                <Button onClick={loadNewScenario} className="flex-1 bg-brand-blue hover:bg-brand-blue/90 text-white gap-2">
                  Next Random Scenario
                  <ArrowRight className="h-4 w-4" />
                </Button>
              </div>
            </CardContent>
          </Card>
        )}
      </div>
    </div>
  )
}
''',
}

for file_path, content in files.items():
    path = ROOT / file_path
    path.parent.mkdir(parents=True, exist_ok=True)
    path.write_text(content.strip() + "\n", encoding="utf-8")
    print(f"Created/updated: {file_path}")

print("\nDone. Practice Lab files created.")
print("Next: run npm run build")
```

```python
from pathlib import Path

ROOT = Path(".")

# -----------------------------
# practice-engine.ts
# -----------------------------
Path("src/lib/practice-engine.ts").write_text(r'''
import {
  practiceScenarios,
  type PracticeRiskLevel,
  type PracticeScenario,
  type PracticeWarningOption,
} from "./practice-scenarios"

const STORAGE_KEY = "nivara_practice_completed_scenarios_v3"

function canUseStorage() {
  return typeof window !== "undefined" && typeof window.localStorage !== "undefined"
}

function validScenarioIdSet() {
  return new Set(practiceScenarios.map((scenario) => scenario.id))
}

export function getCompletedPracticeScenarioIds(): string[] {
  if (!canUseStorage()) return []

  try {
    const raw = window.localStorage.getItem(STORAGE_KEY)
    if (!raw) return []

    const parsed = JSON.parse(raw)
    if (!Array.isArray(parsed)) return []

    const validIds = validScenarioIdSet()
    return parsed.filter((item) => typeof item === "string" && validIds.has(item))
  } catch {
    return []
  }
}

export function saveCompletedPracticeScenarioId(id: string) {
  if (!canUseStorage()) return

  const completed = new Set(getCompletedPracticeScenarioIds())
  completed.add(id)

  window.localStorage.setItem(STORAGE_KEY, JSON.stringify([...completed]))
}

export function resetPracticeSession() {
  if (!canUseStorage()) return
  window.localStorage.removeItem(STORAGE_KEY)
}

export function getRemainingPracticeScenarios(): PracticeScenario[] {
  const completed = new Set(getCompletedPracticeScenarioIds())
  return practiceScenarios.filter((scenario) => !completed.has(scenario.id))
}

export function getRandomPracticeScenario(): PracticeScenario | null {
  const remaining = getRemainingPracticeScenarios()
  if (remaining.length === 0) return null

  const index = Math.floor(Math.random() * remaining.length)
  return remaining[index]
}

export function getVisiblePracticeWarningOptions(
  scenario: PracticeScenario,
  guidedMode: boolean,
): PracticeWarningOption[] {
  if (!guidedMode) return scenario.warningOptions

  const correctSet = new Set(scenario.correctWarningIds)
  const correctOptions = scenario.warningOptions.filter((option) => correctSet.has(option.id))
  const decoyOptions = scenario.warningOptions.filter((option) => !correctSet.has(option.id))

  const chosen: PracticeWarningOption[] = [
    ...correctOptions.slice(0, 3),
    ...decoyOptions.slice(0, 1),
  ]

  const chosenIds = new Set(chosen.map((option) => option.id))

  for (const option of scenario.warningOptions) {
    if (chosen.length >= 4) break
    if (!chosenIds.has(option.id)) {
      chosen.push(option)
      chosenIds.add(option.id)
    }
  }

  return chosen
}

export function getPracticeScore({
  scenario,
  selectedWarningIds,
  selectedRisk,
  scorableWarningIds,
}: {
  scenario: PracticeScenario
  selectedWarningIds: string[]
  selectedRisk: PracticeRiskLevel | null
  scorableWarningIds?: string[]
}) {
  const scorableSet = scorableWarningIds ? new Set(scorableWarningIds) : null

  const correctWarningIds = scorableSet
    ? scenario.correctWarningIds.filter((id) => scorableSet.has(id))
    : scenario.correctWarningIds

  const selectedIds = scorableSet
    ? selectedWarningIds.filter((id) => scorableSet.has(id))
    : selectedWarningIds

  const correctSet = new Set(correctWarningIds)
  const selectedSet = new Set(selectedIds)

  const correctSelected = [...selectedSet].filter((id) => correctSet.has(id)).length
  const extraSelected = [...selectedSet].filter((id) => !correctSet.has(id)).length
  const missed = correctWarningIds.length - correctSelected

  const pointsPerCorrect = correctWarningIds.length > 0 ? 70 / correctWarningIds.length : 0
  const warningScore = Math.max(0, correctSelected * pointsPerCorrect - extraSelected * 4)
  const riskScore = selectedRisk === scenario.riskLevel ? 30 : 0

  let total = Math.max(0, Math.min(100, Math.round(warningScore + riskScore)))

  if (correctSelected > 0 && total < 20) {
    total = 20
  }

  return {
    total,
    correctSelected,
    extraSelected,
    missed,
    riskCorrect: selectedRisk === scenario.riskLevel,
  }
}

export function getPracticeFeedback(
  scenario: PracticeScenario,
  selectedWarningIds: string[],
  scorableWarningIds?: string[],
) {
  const scorableSet = scorableWarningIds ? new Set(scorableWarningIds) : null
  const correctSet = new Set(scenario.correctWarningIds)
  const selectedSet = new Set(selectedWarningIds)

  const options = scorableSet
    ? scenario.warningOptions.filter((option) => scorableSet.has(option.id))
    : scenario.warningOptions

  return {
    correctSelectedOptions: options.filter(
      (option) => correctSet.has(option.id) && selectedSet.has(option.id),
    ),
    missedOptions: options.filter(
      (option) => correctSet.has(option.id) && !selectedSet.has(option.id),
    ),
    extraSelectedOptions: options.filter(
      (option) => !correctSet.has(option.id) && selectedSet.has(option.id),
    ),
  }
}

export function getHiddenCorrectPracticeOptions(
  scenario: PracticeScenario,
  visibleWarningIds: string[],
) {
  const visibleSet = new Set(visibleWarningIds)

  return scenario.warningOptions.filter(
    (option) => scenario.correctWarningIds.includes(option.id) && !visibleSet.has(option.id),
  )
}
'''.strip() + "\n", encoding="utf-8")

# -----------------------------
# AnswerPanel.tsx
# -----------------------------
Path("src/components/practice-lab/AnswerPanel.tsx").write_text(r'''
import { CheckCircle } from "lucide-react"
import { Card, CardContent } from "@/components/ui/card"
import { getVisiblePracticeWarningOptions } from "@/lib/practice-engine"
import type {
  PracticeRiskLevel,
  PracticeScenario,
} from "@/lib/practice-scenarios"

const riskOptions: Array<{
  value: PracticeRiskLevel
  label: string
  helper: string
}> = [
  {
    value: "low",
    label: "Low Risk",
    helper: "Looks mostly normal or expected",
  },
  {
    value: "suspicious",
    label: "Suspicious",
    helper: "Some warning signs are present",
  },
  {
    value: "high",
    label: "High Risk",
    helper: "Multiple strong warning signs",
  },
]

export function AnswerPanel({
  scenario,
  selectedWarningIds,
  selectedRisk,
  submitted,
  guidedMode,
  onToggleWarning,
  onRiskChange,
}: {
  scenario: PracticeScenario
  selectedWarningIds: string[]
  selectedRisk: PracticeRiskLevel | null
  submitted: boolean
  guidedMode: boolean
  onToggleWarning: (id: string) => void
  onRiskChange: (risk: PracticeRiskLevel) => void
}) {
  const visibleOptions = getVisiblePracticeWarningOptions(scenario, guidedMode)

  return (
    <>
      <Card className="mb-6 border-border shadow-sm">
        <CardContent className="p-5">
          <div className="mb-4">
            <p className="text-xs font-bold uppercase tracking-wide text-red-700">
              Step 1
            </p>
            <h2 className="text-lg font-bold text-foreground">
              What signs do you notice?
            </h2>
            <p className="text-sm text-muted-foreground">
              {guidedMode
                ? "Guided Mode shows fewer choices. You are only graded on the choices shown."
                : "Select every sign that clearly appears. Some options are decoys."}
            </p>
          </div>

          <div className="grid gap-2 sm:grid-cols-2">
            {visibleOptions.map((option) => {
              const selected = selectedWarningIds.includes(option.id)

              return (
                <button
                  key={option.id}
                  type="button"
                  onClick={() => onToggleWarning(option.id)}
                  disabled={submitted}
                  className={`rounded-xl border p-3 text-left text-sm transition-all ${
                    selected
                      ? "border-brand-blue bg-brand-blue-light text-brand-blue"
                      : "border-border bg-white hover:bg-muted/30"
                  }`}
                >
                  <div className="flex items-start gap-2">
                    {selected ? (
                      <CheckCircle className="h-4 w-4 mt-0.5 shrink-0" />
                    ) : (
                      <div className="h-4 w-4 rounded-full border-2 border-current mt-0.5 shrink-0" />
                    )}
                    <span className="font-medium">{option.label}</span>
                  </div>
                </button>
              )
            })}
          </div>
        </CardContent>
      </Card>

      <Card className="mb-6 border-border shadow-sm">
        <CardContent className="p-5">
          <div className="mb-4">
            <p className="text-xs font-bold uppercase tracking-wide text-amber-700">
              Step 2
            </p>
            <h2 className="text-lg font-bold text-foreground">
              What risk level would you choose?
            </h2>
            <p className="text-sm text-muted-foreground">
              This helps you practice judgment, not just memorization.
            </p>
          </div>

          <div className="grid gap-2 sm:grid-cols-3">
            {riskOptions.map((option) => {
              const selected = selectedRisk === option.value

              return (
                <button
                  key={option.value}
                  type="button"
                  onClick={() => !submitted && onRiskChange(option.value)}
                  disabled={submitted}
                  className={`rounded-xl border p-4 text-left transition-all ${
                    selected
                      ? "border-amber-300 bg-amber-50 text-amber-800"
                      : "border-border bg-white hover:bg-muted/30"
                  }`}
                >
                  <p className="text-sm font-bold">{option.label}</p>
                  <p className="mt-1 text-xs text-muted-foreground">
                    {option.helper}
                  </p>
                </button>
              )
            })}
          </div>
        </CardContent>
      </Card>
    </>
  )
}
'''.strip() + "\n", encoding="utf-8")

# -----------------------------
# FeedbackPanel.tsx
# -----------------------------
Path("src/components/practice-lab/FeedbackPanel.tsx").write_text(r'''
import { Link } from "react-router-dom"
import {
  ArrowRight,
  BookOpen,
  CheckCircle,
  Lightbulb,
  Shield,
  Sparkles,
  XCircle,
} from "lucide-react"
import { Button } from "@/components/ui/button"
import { Card, CardContent } from "@/components/ui/card"
import {
  getHiddenCorrectPracticeOptions,
  getPracticeFeedback,
  getPracticeScore,
  getVisiblePracticeWarningOptions,
} from "@/lib/practice-engine"
import type {
  PracticeRiskLevel,
  PracticeScenario,
  PracticeWarningOption,
} from "@/lib/practice-scenarios"

function FeedbackList({
  title,
  tone,
  items,
  empty,
  extraMode = false,
}: {
  title: string
  tone: "green" | "red" | "amber" | "blue"
  items: PracticeWarningOption[]
  empty: string
  extraMode?: boolean
}) {
  const styles = {
    green: "border-emerald-200 bg-emerald-50 text-emerald-700",
    red: "border-red-200 bg-red-50 text-red-700",
    amber: "border-amber-200 bg-amber-50 text-amber-700",
    blue: "border-blue-200 bg-blue-50 text-blue-700",
  }

  const Icon =
    tone === "green"
      ? CheckCircle
      : tone === "red"
        ? XCircle
        : tone === "blue"
          ? Lightbulb
          : Sparkles

  return (
    <div className={`rounded-2xl border p-4 ${styles[tone]}`}>
      <p className="text-sm font-bold mb-3">{title}</p>

      {items.length === 0 ? (
        <p className="text-sm text-muted-foreground">{empty}</p>
      ) : (
        <div className="space-y-2">
          {items.map((item) => (
            <div key={item.id} className="rounded-xl border border-white/70 bg-white p-3">
              <div className="flex items-start gap-2">
                <Icon className="h-4 w-4 mt-0.5 shrink-0" />
                <div>
                  <p className="text-sm font-bold text-foreground">{item.label}</p>
                  <p className="mt-1 text-xs text-muted-foreground leading-relaxed">
                    {extraMode
                      ? "This can matter in other scams, but this situation does not clearly show it."
                      : item.why}
                  </p>
                </div>
              </div>
            </div>
          ))}
        </div>
      )}
    </div>
  )
}

export function FeedbackPanel({
  scenario,
  selectedWarningIds,
  selectedRisk,
  guidedMode,
  onNextScenario,
}: {
  scenario: PracticeScenario
  selectedWarningIds: string[]
  selectedRisk: PracticeRiskLevel | null
  guidedMode: boolean
  onNextScenario: () => void
}) {
  const visibleOptions = getVisiblePracticeWarningOptions(scenario, guidedMode)
  const visibleWarningIds = visibleOptions.map((option) => option.id)

  const score = getPracticeScore({
    scenario,
    selectedWarningIds,
    selectedRisk,
    scorableWarningIds: guidedMode ? visibleWarningIds : undefined,
  })

  const feedback = getPracticeFeedback(
    scenario,
    selectedWarningIds,
    guidedMode ? visibleWarningIds : undefined,
  )

  const hiddenCorrectOptions = guidedMode
    ? getHiddenCorrectPracticeOptions(scenario, visibleWarningIds)
    : []

  return (
    <Card className="mb-8 border-brand-blue/20 bg-brand-blue-light/20 shadow-sm">
      <CardContent className="p-5">
        <div className="text-center mb-5">
          <Sparkles className="mx-auto mb-2 h-7 w-7 text-brand-blue" />
          <p className="text-3xl font-extrabold text-brand-blue">{score.total}%</p>
          <p className="text-sm text-muted-foreground">
            {score.total === 100
              ? "Excellent. You spotted the shown pattern clearly."
              : score.total >= 70
                ? "Good work. Review the details below."
                : "Good practice. This one had some tricky signals."}
          </p>
        </div>

        <div className="grid gap-3 sm:grid-cols-2 mb-5">
          <div className="rounded-xl border border-border bg-white p-4">
            <p className="text-xs font-bold text-muted-foreground uppercase tracking-wide">
              Scenario Type
            </p>
            <p className="text-sm font-bold text-foreground mt-1">
              {scenario.hiddenCategory}
            </p>
          </div>

          <div className="rounded-xl border border-border bg-white p-4">
            <p className="text-xs font-bold text-muted-foreground uppercase tracking-wide">
              Risk Level
            </p>
            <p className="text-sm font-bold text-foreground mt-1">
              {scenario.riskLevel.toUpperCase()}
            </p>
            {!score.riskCorrect && (
              <p className="mt-1 text-xs text-red-600">
                Your selected risk level was different.
              </p>
            )}
          </div>
        </div>

        <div className="grid gap-3 mb-5">
          <FeedbackList
            title="Correctly noticed"
            tone="green"
            items={feedback.correctSelectedOptions}
            empty="You did not select any correct signs this time."
          />

          <FeedbackList
            title={guidedMode ? "Missed signs from shown choices" : "Missed signs"}
            tone="red"
            items={feedback.missedOptions}
            empty="No missed signs from the choices shown. Nice work."
          />

          <FeedbackList
            title="Extra signs selected"
            tone="amber"
            items={feedback.extraSelectedOptions}
            empty="No extra signs selected."
            extraMode
          />

          {guidedMode && hiddenCorrectOptions.length > 0 && (
            <FeedbackList
              title="More signs to remember"
              tone="blue"
              items={hiddenCorrectOptions}
              empty="No extra learning notes."
            />
          )}
        </div>

        <div className="rounded-xl border border-border bg-white p-4 mb-4">
          <p className="text-xs font-bold text-muted-foreground uppercase tracking-wide">
            Explanation
          </p>
          <p className="mt-2 text-sm text-foreground leading-relaxed">
            {scenario.explanation}
          </p>
        </div>

        <div className="rounded-xl border border-emerald-200 bg-emerald-50 p-4 mb-5">
          <div className="flex items-start gap-2">
            <Shield className="h-4 w-4 text-emerald-600 mt-0.5 shrink-0" />
            <div>
              <p className="text-sm font-bold text-emerald-700">Safer move</p>
              <p className="mt-1 text-sm text-foreground leading-relaxed">
                {scenario.safeAction}
              </p>
            </div>
          </div>
        </div>

        <div className="flex flex-col sm:flex-row gap-2">
          <Link to={`/learning/${scenario.relatedLessonId}`} className="flex-1">
            <Button
              variant="outline"
              className="w-full gap-2 border-brand-blue/30 text-brand-blue hover:bg-brand-blue-light"
            >
              <BookOpen className="h-4 w-4" />
              Learn This Pattern
            </Button>
          </Link>

          <Button
            onClick={onNextScenario}
            className="flex-1 bg-brand-blue hover:bg-brand-blue/90 text-white gap-2"
          >
            Next Random Scenario
            <ArrowRight className="h-4 w-4" />
          </Button>
        </div>
      </CardContent>
    </Card>
  )
}
'''.strip() + "\n", encoding="utf-8")

# -----------------------------
# Patch ScamLab.tsx to pass guidedMode into FeedbackPanel
# -----------------------------
path = Path("src/pages/ScamLab.tsx")
text = path.read_text(encoding="utf-8")

text = text.replace(
'''          <FeedbackPanel
            scenario={scenario}
            selectedWarningIds={selectedWarningIds}
            selectedRisk={selectedRisk}
            onNextScenario={loadNewScenario}
          />''',
'''          <FeedbackPanel
            scenario={scenario}
            selectedWarningIds={selectedWarningIds}
            selectedRisk={selectedRisk}
            guidedMode={guidedMode}
            onNextScenario={loadNewScenario}
          />'''
)

path.write_text(text, encoding="utf-8")

print("Guided Mode grading fixed.")
print("Now run: npm run build")
```

```python
from pathlib import Path
import re

ROOT = Path(".")
SRC = ROOT / "src"

errors = []
warnings = []
passes = []

def read(path):
    return Path(path).read_text(encoding="utf-8", errors="ignore")

def add_error(msg):
    errors.append(msg)

def add_warning(msg):
    warnings.append(msg)

def add_pass(msg):
    passes.append(msg)

print("NIVARA BRUTAL STRESS TEST STARTING...\n")

# --------------------------------------------------
# 1. Required file structure
# --------------------------------------------------
required_files = [
    "src/App.tsx",
    "src/components/Navbar.tsx",
    "src/components/Footer.tsx",
    "src/pages/ScamLab.tsx",
    "src/pages/ScamCheck.tsx",
    "src/pages/Results.tsx",
    "src/pages/LearningCenter.tsx",
    "src/pages/LessonDetail.tsx",
    "src/pages/Pricing.tsx",
    "src/pages/PrivacySafety.tsx",
    "src/pages/ReportRecover.tsx",
    "src/lib/scam-lab.ts",
    "src/lib/practice-engine.ts",
    "src/lib/practice-scenarios.ts",
    "src/components/practice-lab/ScenarioCard.tsx",
    "src/components/practice-lab/AnswerPanel.tsx",
    "src/components/practice-lab/FeedbackPanel.tsx",
]

for f in required_files:
    if not Path(f).exists():
        add_error(f"Missing required file: {f}")
    else:
        add_pass(f"File exists: {f}")

# --------------------------------------------------
# 2. Doxxing / personal info scan
# --------------------------------------------------
banned_terms = [
    "Archana",
    "archana",
    "Renuka",
    "renuka",
    "archkaushal",
    "Hoshiarpur",
    "Tarrant County College",
    "University of Texas at Arlington",
    "UTA",
    "TCC",
    "Midlothian",
    "Sulphur Springs",
]

for path in SRC.rglob("*"):
    if path.suffix not in {".ts", ".tsx", ".js", ".jsx", ".md"}:
        continue

    text = read(path)

    for term in banned_terms:
        if term in text:
            add_error(f"Possible personal/private detail found in {path}: {term}")

if not any("personal/private detail" in e for e in errors):
    add_pass("No banned personal/private names found.")

# --------------------------------------------------
# 3. Old visible label scan
# --------------------------------------------------
for path in SRC.rglob("*"):
    if path.suffix not in {".ts", ".tsx", ".js", ".jsx", ".md"}:
        continue

    text = read(path)

    if "Scam Lab" in text:
        add_warning(f"Old visible label may remain in {path}: Scam Lab")

# --------------------------------------------------
# 4. Route consistency
# --------------------------------------------------
app_text = read("src/App.tsx") if Path("src/App.tsx").exists() else ""
navbar_text = read("src/components/Navbar.tsx") if Path("src/components/Navbar.tsx").exists() else ""
footer_text = read("src/components/Footer.tsx") if Path("src/components/Footer.tsx").exists() else ""

routes = set(re.findall(r'path=["\']([^"\']+)["\']', app_text))
links = set(re.findall(r'(?:to|href):?\s*=?\s*["\']([^"\']+)["\']', navbar_text + "\n" + footer_text))

required_routes = {
    "/",
    "/check",
    "/results",
    "/scam-lab",
    "/learning",
    "/learning/:id",
    "/report",
    "/privacy",
    "/pricing",
    "/early-access",
}

for route in required_routes:
    if route not in routes:
        add_error(f"Missing route in App.tsx: {route}")
    else:
        add_pass(f"Route exists: {route}")

required_nav_links = {
    "/check",
    "/scam-lab",
    "/learning",
    "/report",
    "/privacy",
    "/pricing",
}

for link in required_nav_links:
    if link not in links:
        add_error(f"Missing navbar/footer link: {link}")
    else:
        add_pass(f"Link exists: {link}")

# --------------------------------------------------
# 5. Scenario bank parsing
# --------------------------------------------------
scenario_file = Path("src/lib/scam-lab.ts")

if not scenario_file.exists():
    add_error("Missing scenario file: src/lib/scam-lab.ts")
else:
    scenario_text = read(scenario_file)

    # Warning IDs from warningLibrary
    warning_ids = set(re.findall(r'\n\s*([a-zA-Z0-9_]+):\s*\{\s*\n\s*id:\s*"([^"]+)"', scenario_text))
    warning_key_ids = {pair[0] for pair in warning_ids}
    warning_value_ids = {pair[1] for pair in warning_ids}
    known_warning_ids = warning_key_ids | warning_value_ids

    if len(known_warning_ids) < 20:
        add_warning(f"Only {len(known_warning_ids)} warning IDs detected. Check warningLibrary parsing.")
    else:
        add_pass(f"Warning library detected: {len(known_warning_ids)} warning IDs")

    # Scenario IDs
    scenario_ids = re.findall(r'id:\s*"([^"]+)"', scenario_text)
    scenario_ids = [x for x in scenario_ids if x.startswith(("launch-", "hard-", "practice-"))]

    duplicate_scenario_ids = sorted({x for x in scenario_ids if scenario_ids.count(x) > 1})
    if duplicate_scenario_ids:
        add_error(f"Duplicate scenario IDs found: {duplicate_scenario_ids}")
    else:
        add_pass("No duplicate scenario IDs found.")

    scenario_count = len(scenario_ids)
    if scenario_count < 25:
        add_warning(f"Only {scenario_count} scenarios found. For launch, 30+ is better.")
    else:
        add_pass(f"Scenario count looks okay: {scenario_count}")

    # Parse rough scenario blocks
    blocks = re.findall(
        r'\{\s*id:\s*"([^"]+)"[\s\S]*?relatedLessonLabel:\s*"[^"]+",?\s*\}',
        scenario_text
    )

    if len(blocks) < scenario_count:
        add_warning("Scenario block parser found fewer blocks than IDs. Regex may be imperfect.")
    else:
        add_pass("Scenario blocks detected.")

    # Full block extraction
    raw_blocks = re.findall(
        r'\{\s*id:\s*"([^"]+)"[\s\S]*?decoyWarningIds:\s*\[([^\]]*)\][\s\S]*?relatedLessonLabel:\s*"([^"]+)",?\s*\}',
        scenario_text
    )

    for sid, decoy_raw, label in raw_blocks:
        pass

    # Better block parser by splitting at ids
    split_blocks = re.split(r'\n\s*\{\s*id:\s*"', scenario_text)
    scenario_blocks = []

    for chunk in split_blocks[1:]:
        sid_match = re.match(r'([^"]+)"', chunk)
        if not sid_match:
            continue
        sid = sid_match.group(1)
        if sid.startswith(("launch-", "hard-", "practice-")):
            scenario_blocks.append((sid, chunk))

    required_fields = [
        "format:",
        "title:",
        "hiddenCategory:",
        "difficulty:",
        "riskLevel:",
        "content:",
        "correctWarningIds:",
        "decoyWarningIds:",
        "explanation:",
        "safeAction:",
        "relatedLessonId:",
        "relatedLessonLabel:",
    ]

    risky_correct_signs = {
        "suspicious_link",
        "credential_login",
        "urgency",
        "threat",
        "personal_info",
        "payment_request",
        "gift_cards",
        "secrecy",
        "avoid_official",
        "verification_code",
        "too_good",
        "unusual_payment",
        "remote_access",
        "malware_download",
        "sensitive_docs",
        "asks_to_call_number",
        "fake_invoice",
        "overpayment",
    }

    expected_text_clues = {
        "suspicious_link": ["http", ".net", ".com", ".org", "link", "domain", "website", "QR"],
        "credential_login": ["sign in", "login", "log in", "verify", "account", "credentials"],
        "urgency": ["today", "immediately", "before", "expires", "urgent", "now", "tonight", "deadline", "6 PM", "10 minutes"],
        "threat": ["blocked", "locked", "suspended", "disabled", "removed", "penalty", "shut", "disconnect", "restriction", "interruption"],
        "personal_info": ["name", "address", "card", "date of birth", "insurance", "SSN", "ID", "details"],
        "payment_request": ["pay", "payment", "fee", "deposit", "wire", "Zelle", "Cash App", "crypto", "card", "checkout", "donations"],
        "gift_cards": ["gift card", "Steam", "Apple", "codes", "card codes"],
        "secrecy": ["do not tell", "keep this", "do not involve", "between us", "private"],
        "avoid_official": ["do not call", "do not contact", "cannot access", "do not involve", "do not try logging in"],
        "verification_code": ["verification code", "code", "one-time", "passcode"],
        "too_good": ["selected", "discount", "higher", "returns", "$36/hour", "below market", "half", "free", "eligible"],
        "unusual_payment": ["Zelle", "Cash App", "crypto", "wire", "bank transfer", "Apple Pay", "personal payment", "gift card"],
        "remote_access": ["remote", "session", "screen", "technician", "support tool"],
        "malware_download": ["download", "install", ".download", "tool"],
        "sensitive_docs": ["SSN", "photo ID", "passport", "identity", "documents", "bank name", "tax"],
        "asks_to_call_number": ["call", "billing desk", "number"],
        "fake_invoice": ["invoice", "charge", "renewed", "billing", "past due"],
        "wrong_number": ["wrong chat", "wrong number"],
        "overpayment": ["extra", "difference", "overpaid", "sent extra"],
        "emotional_manipulation": ["trust", "trouble", "family", "help", "scared", "need help"],
        "qr_code_risk": ["QR"],
        "ai_or_fake_media": ["voice", "AI", "fake", "video", "sounds like"],
        "payment_change": ["banking details changed", "new account", "changed", "new payment"],
        "out_of_context": ["unexpected", "not expected", "does not match", "out of context"],
    }

    for sid, block in scenario_blocks:
        for field in required_fields:
            if field not in block:
                add_error(f"{sid}: missing required field {field}")

        correct_match = re.search(r'correctWarningIds:\s*\[([^\]]*)\]', block)
        decoy_match = re.search(r'decoyWarningIds:\s*\[([^\]]*)\]', block)
        body_match = re.search(r'body:\s*"([^"]+)"', block)

        if not correct_match:
            add_error(f"{sid}: missing correctWarningIds array")
            continue

        if not decoy_match:
            add_error(f"{sid}: missing decoyWarningIds array")
            continue

        correct_ids = re.findall(r'"([^"]+)"', correct_match.group(1))
        decoy_ids = re.findall(r'"([^"]+)"', decoy_match.group(1))
        body = body_match.group(1) if body_match else ""

        if len(correct_ids) < 2:
            add_warning(f"{sid}: has fewer than 2 correct warning signs.")

        if len(decoy_ids) < 2:
            add_warning(f"{sid}: has fewer than 2 decoy warning signs.")

        overlap = set(correct_ids) & set(decoy_ids)
        if overlap:
            add_error(f"{sid}: warning IDs appear in both correct and decoy: {sorted(overlap)}")

        for wid in correct_ids + decoy_ids:
            if wid not in known_warning_ids:
                add_error(f"{sid}: unknown warning ID used: {wid}")

        # Text clue checks
        for wid in correct_ids:
            clues = expected_text_clues.get(wid)
            if not clues:
                continue

            if not any(clue.lower() in body.lower() for clue in clues):
                add_warning(f"{sid}: correct warning '{wid}' may not be clearly visible in body.")

        # Risk sanity
        risk_match = re.search(r'riskLevel:\s*"([^"]+)"', block)
        risk = risk_match.group(1) if risk_match else ""

        if risk == "low":
            alarming = set(correct_ids) & risky_correct_signs
            if alarming:
                add_warning(f"{sid}: low-risk scenario includes risky warning signs: {sorted(alarming)}")

        if risk == "high" and len(correct_ids) < 3:
            add_warning(f"{sid}: high-risk scenario has fewer than 3 correct signs.")

        if body and len(body) < 45:
            add_warning(f"{sid}: scenario body may be too short to teach well.")

        if body and len(body) > 420:
            add_warning(f"{sid}: scenario body may be too long for mobile.")

# --------------------------------------------------
# 6. Redaction logic sanity
# --------------------------------------------------
analyzer_path = Path("src/lib/scam-analyzer.ts")

if analyzer_path.exists():
    analyzer_text = read(analyzer_path)

    needed_redactions = [
        "[EMAIL HIDDEN]",
        "[PHONE HIDDEN]",
        "[ADDRESS HIDDEN]",
    ]

    for token in needed_redactions:
        if token not in analyzer_text:
            add_warning(f"Redaction token may be missing: {token}")
        else:
            add_pass(f"Redaction token found: {token}")

    if "[NAME HIDDEN]" not in analyzer_text:
        add_warning("Name redaction token may be missing: [NAME HIDDEN]")

    if "[CODE HIDDEN]" not in analyzer_text and "verification" not in analyzer_text.lower():
        add_warning("Verification code redaction may be weak or missing.")
else:
    add_warning("scam-analyzer.ts not found.")

# --------------------------------------------------
# 7. Safety disclaimer scan
# --------------------------------------------------
all_src = ""
for path in SRC.rglob("*"):
    if path.suffix in {".ts", ".tsx", ".js", ".jsx"}:
        all_src += read(path) + "\n"

disclaimer_checks = [
    "does not guarantee scam detection",
    "not a substitute",
    "Always verify independently",
    "Educational",
]

for phrase in disclaimer_checks:
    if phrase not in all_src:
        add_warning(f"Disclaimer phrase may be missing: {phrase}")
    else:
        add_pass(f"Disclaimer phrase found: {phrase}")

# --------------------------------------------------
# 8. Practice Lab state behavior scan
# --------------------------------------------------
practice_engine = Path("src/lib/practice-engine.ts")
scamlab_page = Path("src/pages/ScamLab.tsx")

if practice_engine.exists():
    engine_text = read(practice_engine)

    required_engine_functions = [
        "getRandomPracticeScenario",
        "saveCompletedPracticeScenarioId",
        "resetPracticeSession",
        "getVisiblePracticeWarningOptions",
        "getPracticeScore",
        "getPracticeFeedback",
    ]

    for fn in required_engine_functions:
        if fn not in engine_text:
            add_error(f"Practice engine missing function: {fn}")
        else:
            add_pass(f"Practice engine function exists: {fn}")

    if "localStorage" not in engine_text:
        add_warning("Practice engine may not persist non-repeat scenario state.")
    else:
        add_pass("Practice engine uses localStorage.")
else:
    add_error("Missing practice-engine.ts")

if scamlab_page.exists():
    page_text = read(scamlab_page)

    if "GUIDED_MODE_KEY" not in page_text:
        add_warning("Guided Mode may not persist after refresh.")
    else:
        add_pass("Guided Mode persistence key found.")

    if "scrollIntoView" not in page_text:
        add_warning("Page may not scroll after submit/next scenario.")
    else:
        add_pass("Auto-scroll behavior found.")

# --------------------------------------------------
# 9. Final report
# --------------------------------------------------
print("\nBRUTAL STRESS TEST RESULTS")
print("=" * 60)

if errors:
    print("\nERRORS:")
    for e in errors:
        print("❌", e)
else:
    print("\n✅ No critical errors found.")

if warnings:
    print("\nWARNINGS:")
    for w in warnings:
        print("⚠️", w)
else:
    print("\n✅ No warnings found.")

print("\nPASSES:")
for p in passes[:40]:
    print("✅", p)

if len(passes) > 40:
    print(f"✅ ...and {len(passes) - 40} more passes")

print("\nSUMMARY")
print(f"Critical errors: {len(errors)}")
print(f"Warnings: {len(warnings)}")
print(f"Passes: {len(passes)}")

if errors:
    print("\nRESULT: FAIL — fix critical errors before manual testing.")
elif len(warnings) > 15:
    print("\nRESULT: PASS WITH MANY WARNINGS — review warnings before launch.")
else:
    print("\nRESULT: STRONG PASS — ready for browser/manual stress testing.")
```
